## Handle missing values

### Subtask:
Address missing values in both the training and testing datasets using appropriate strategies (e.g., imputation, dropping columns/rows) based on the nature and extent of missingness.


**Reasoning**:
Identify and drop columns with a high percentage of missing values (above 50%) in both dataframes as they are unlikely to provide useful information.



```python
missing_threshold = 50

cols_to_drop_train = missing_train_percentage[missing_train_percentage > missing_threshold].index
train_df = train_df.drop(columns=cols_to_drop_train)

cols_to_drop_test = missing_test_percentage[missing_test_percentage > missing_threshold].index
test_df = test_df.drop(columns=cols_to_drop_test)

display(train_df.head())
display(test_df.head())
```

**Reasoning**:
Impute missing values in numerical and categorical columns using appropriate strategies based on the remaining missing values after dropping high-missing columns. For numerical columns, use the median, and for categorical columns, use the mode or a 'None' category where appropriate based on the data description.



```python
for col in train_df.columns:
    if train_df[col].isnull().any():
        if train_df[col].dtype in ['int64', 'float64']:
            train_df[col] = train_df[col].fillna(train_df[col].median())
        else:
            train_df[col] = train_df[col].fillna(train_df[col].mode()[0])

for col in test_df.columns:
    if test_df[col].isnull().any():
        if test_df[col].dtype in ['int64', 'float64']:
            test_df[col] = test_df[col].fillna(test_df[col].median())
        else:
            test_df[col] = test_df[col].fillna(test_df[col].mode()[0])

display(train_df.isnull().sum().sum())
display(test_df.isnull().sum().sum())
```

## Implement hyperparameter tuning

### Subtask:
Use a tuning technique (e.g., GridSearchCV or RandomizedSearchCV) with the chosen cross-validation strategy and hyperparameter grid to find the best hyperparameters.

**Reasoning**:
Import GridSearchCV, create a LightGBM regressor, instantiate GridSearchCV with the defined parameter grid and KFold object, and fit it to the training data to find the best hyperparameters.

```python
from sklearn.model_selection import GridSearchCV
import lightgbm as lgb

lgbm = lgb.LGBMRegressor(random_state=42)

grid_search = GridSearchCV(estimator=lgbm, param_grid=param_grid, cv=kf, scoring='neg_root_mean_squared_error')

grid_search.fit(X_train, y_train)

print("Best hyperparameters found: ", grid_search.best_params_)
```

## Evaluate the tuned model

### Subtask:
Train the LightGBM model with the best hyperparameters found and evaluate its performance on the training data using cross-validation.

**Reasoning**:
Train the LightGBM model with the best hyperparameters found and evaluate its performance on the training data using cross-validation.

```python
from sklearn.model_selection import cross_val_score
import numpy as np

best_params = grid_search.best_params_

lgbm_best = lgb.LGBMRegressor(**best_params, random_state=42)

cv_scores = cross_val_score(lgbm_best, X_train, y_train, cv=kf, scoring='neg_root_mean_squared_error')

rmse_scores = -cv_scores
mean_rmse = np.mean(rmse_scores)
std_rmse = np.std(rmse_scores)

print(f"Cross-validated RMSE: {mean_rmse:.4f} +/- {std_rmse:.4f}")
```
